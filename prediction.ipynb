```python
cp.columns = ['Country_Name', 'Country_Code', 'Indicator_Name', 'Indicator_Code',
       '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013']
```

```python
###After Replacing space in between the columns
cp.columns
```

```python
def population():
    cname = input("Enter the Country Name:")  ####Gettting the country name from the user
    cname = cname.capitalize()                ####Capitalizing first letter(All the countries name starts with capsin the dataset)
    cname1 = cp[cp.Country_Name == cname]     ###Fetching the given country
    for i in cp.Country_Name:                 ###to check with all the country name
        if cname  == i:                       ###Checking wheatrher the country namme entered is present
            X1 = cname1.iloc[:, 4:].values
            y1 = cname1.columns[4:]
        
            l1 = y1
            l1 = list(map(int, l1))
    for j in X1:
        n1 = list(j)
        n1 = list(map(int, n1))
    df = pd.DataFrame({                 ##Creating desired data frame
            "Year": np.array(l1),
            "Population":np.array(n1)})
    X = df.iloc[:, 1].values         ###Assigning Years to X
    Y = df.iloc[:, 0].values         ###Assigning population to Y
    X = np.reshape(X, (-1, 1))       ###Reshaping to 2D array
    Y = np.reshape(Y, (-1,1))        ###Reshaping to 2D array
    
    from sklearn.cross_validation import train_test_split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
    
    from sklearn.linear_model import LinearRegression
    regressor = LinearRegression()  ###Assigning Regression model to the variable
    regressor.fit(X_train,Y_train)  ###Traingng the train datas with machine
    
    pred = regressor.predict(X_test) ###predicting the test data
    
    plt.scatter(X_test,Y_test,color = 'red')   ###Visualizing actual test data
    plt.plot(X_test,pred,color = 'purple')     ###Visualizing predicted data    
    plt.title('Population Prediction for ' + cname) ###title for the chart
    plt.show()
   
```

```python
population()
```
